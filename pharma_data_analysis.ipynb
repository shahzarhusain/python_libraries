```python
df = pd.read_csv('Pfizer_1.csv')
```

```python
# melt is kind of anti pivoting, this will bring the col as category to each row
pd.melt(df, id_vars = ['Date', 'Drug_Name', 'Parameter']).sort_values(by = ['Date', 'Drug_Name', 'Parameter'])
```

```python
# let's save our melted data with proper names
melted_df = pd.melt(df, id_vars= ['Date','Drug_Name','Parameter'],
                    var_name = 'Time',
                    value_name= 'Readings')
melted_df.head()
```

```python
# let's try to group our data
melted_df.groupby(['Date','Drug_Name','Parameter'])['Readings'].mean()
```

```python
# now, just to practive the pivot, let's try to revert back to the melted data
melted_df.pivot(index = ['Date', 'Drug_Name', 'Parameter'],
                columns = 'Time',
                values = 'Readings')
```

```python
pivoted_data = melted_df.pivot(index = ['Date', 'Drug_Name', 'Parameter'],
                columns = 'Time',
                values = 'Readings').reset_index()
```

```python
pivoted_data.columns.name = None
pivoted_data.head()
```

```python
  # now we will try to make our data more readable by pivoting the Parameters
data_tidy = melted_df.pivot(index = ['Date','Drug_Name', 'Time'], columns = 'Parameter', values = 'Readings').reset_index()
data_tidy.columns.name = None
data_tidy.head()
```

```python
# can we pivot with one col?
data_tidy.pivot(index = ['Drug_Name'], columns = 'Date', values = 'Temperature')
```

```python
data_tidy.head()
```

```python
# as we can see that the pivot function is not able to handle the duplicate values and also it does not have a way to aggregate the values.
#Therefore, we have a function called pivot_table to tackle with this
data_tidy.pivot_table(index = 'Drug_Name', columns = 'Date', values = 'Temperature', aggfunc = np.mean)
```

```python
# for each drug we can get the average pressure and temperature for each day
data_tidy.pivot_table(index = ['Drug_Name'], columns = 'Date', values = ['Temperature','Pressure'], aggfunc = np.mean)
```

```python
# Similary, we can get the min pressure and temperature for each day for each drug
data_tidy.pivot_table(index = ['Drug_Name'], columns = 'Date', values = ['Temperature','Pressure'], aggfunc = np.min)
```

```python
df['2:30:00']
```

```python
# let's try to fill the values with 0
df['2:30:00'].fillna(0)
```

```python
# to fill the vlaues with the mean value
df['2:30:00'].fillna(df['2:30:00'].mean())
```

```python
# now, as we have already simplified and transformed our data. Let's try to fill the null values
data_tidy.head()
```

```python
data_tidy.isnull().sum()
```

```python
data_tidy[data_tidy['Pressure'].isnull()]
```

```python
# let's create a function first to calculate the mean temprature and pressure column
def avg_values(x):
  x['Avg_temprature'] = x['Temperature'].mean()
  x['Avg_pressure'] = x['Pressure'].mean()
  return x
```

```python
data_tidy.groupby(['Drug_Name'])[['Temperature','Pressure']].mean()
```

```python
data_tidy = data_tidy.groupby('Drug_Name').apply(avg_values)
```

```python
# now, as we have extracted the average values, we will fill our null values with them
data_tidy['Temperature'].fillna(data_tidy['Avg_temprature'], inplace = True)
data_tidy['Pressure'].fillna(data_tidy['Avg_pressure'], inplace = True)
```

```python
data_tidy.isnull().sum()
```

```python
data_tidy.loc[[5,15,21,28]]
# Voila! we got our null values filled without any loss of data
```

# PD.cut

```python
# We can use this functinality to categorize our NUMERICAL data
# for example - categorizing the temperature in low, med high
```

```python
# Let's see the range of our temperature data before creating the bins
print(data_tidy['Temperature'].min(), data_tidy['Temperature'].max())
```

```python
temp_points = [5,20,35,50,60]   # this will work as our bin i.e. 8-20, 20-35, 35-50, 50-60
temp_lables = ['low','med','high','very_high']

data_tidy['temp_cat'] = pd.cut(data_tidy['Temperature'], bins=temp_points, labels = temp_lables)
```
